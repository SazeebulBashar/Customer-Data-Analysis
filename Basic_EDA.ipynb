# Basic Exploratory Data Analysis (EDA)

This notebook performs basic exploratory data analysis on the Customer Shopping Behavior dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visualizations
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('Datasets/customer_shopping_behavior.csv')
print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
```

```python
# Display first 5 rows
df.head()
```

```python
# Get basic info about the dataset
df.info()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Summary for categorical columns
df.describe(include='object')
```

```python
# Check for missing values
print('Missing values per column:')
df.isnull().sum()
```

```python
# Check unique values in categorical columns
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f'{col}: {df[col].nunique()} unique values')
    print(f'Values: {df[col].unique()}\n')
```

```python
# Distribution of numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols].hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Boxplot for Purchase Amount
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Purchase Amount (USD)'])
plt.title('Boxplot of Purchase Amount (USD)')
plt.show()
```

```python
# Count plot for Category
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df, order=df['Category'].value_counts().index)
plt.title('Count of Items by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count plot for Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df)
plt.title('Count by Gender')
plt.show()
```

```python
# Correlation matrix for numerical columns
plt.figure(figsize=(10, 8))
correlation = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Count plot for Season
plt.figure(figsize=(8, 5))
sns.countplot(x='Season', data=df, order=df['Season'].value_counts().index)
plt.title('Count by Season')
plt.show()
```

```python
# Basic insights
print('Basic Insights:')
print(f'1. Total records: {df.shape[0]}')
print(f'2. Total features: {df.shape[1]}')
print(f'3. Missing values: {df.isnull().sum().sum()}')
print(f'4. Average age: {df["Age"].mean():.2f} years')
print(f'5. Average purchase amount: ${df["Purchase Amount (USD)"].mean():.2f}')
print(f'6. Most common category: {df["Category"].mode()[0]}')
print(f'7. Most common payment method: {df["Payment Method"].mode()[0]}')
```
